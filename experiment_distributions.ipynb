## Cauchi distribution

```python
def logp_cauchy(x):
    return -torch.log(1 + (x ** 2).sum(dim=-1))
example_log_pdf_1 = lambda x: -torch.log(1 + (x ** 2).sum(dim=-1))
```

```python
test = UBVI(logp_cauchy, Gaussian(1), num_opt_steps=1000, n_samples=100000, n_init=100, init_inflation=100)
```

```python
hellinger_value = []
weights_list = []
samples_list = []
for i in range(1, 17):
    test.build(i)
    weights_list.append(test.get_weights().detach().numpy())
    samples_list.append(test._sample_g(100000).detach().numpy().flatten())
    res = test._hellsq_estimate()
    hellinger_value.append(res.data.numpy().reshape((1))[0])
```

```python
weights_list
```

```python
fig, ax = plt.subplots(4, 4, figsize=(16, 16))


# fig.title('Boosting Cauchy with gaussians')
for i in range(4):
    for j in range(4):
        ax[i, j].hist(samples_list[4 * i + j], density=True, bins=120, range=[-10, 10])
        grid = np.linspace(-10, 10, 100)
        ax[i, j].plot(grid, (1 / np.pi) / (1 + grid ** 2))
        ax[i, j].title.set_text('Boosting Cauchy with {} gaussians'.format(4 * i + j + 1))
        ax[i, j].set_ylabel(r'$p$')
        ax[i, j].set_xlabel(r'$x$')

plt.show()
```

```python
plt.figure(figsize=(15, 8))
x = [i for i in range(1, 17)]
plt.plot(x, hellinger_value)
plt.xlabel('Number of components')
plt.ylabel('Hellinger distance')
plt.title("Hellinger distance vs number of components")
```

## 2d Gaussian distribution with non-diagonal covariance matrix

```python
logp_2d_gaussian = lambda x: -(x[:, 0] - x[:, 1]) ** 2 / 2 - (x[:, 1] + x[:, 0]) ** 2 / (2 * 2 ** 2)
```

```python
test = UBVI(logp_2d_gaussian, Gaussian(2), num_opt_steps=5000, n_samples=1000, n_init=500, init_inflation=100)
```

```python
test.build(10)
```

```python
samples = test._sample_g(100000).detach().numpy()

plt.figure(figsize=(10, 10))

plt.scatter(samples[:, 0], samples[:, 1], c=(0, 0, 1, 0.05))


x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
X, Y = np.meshgrid(x, y)
plt.contour(X, Y, -(X - Y) ** 2 / 2 - (X + Y) ** 2 / (2 * 2 ** 2), [-10, -5, -1], colors='k')
plt.xlim([-7, 7])
plt.ylim([-7, 7])

# grid = np.linspace(-10, 10, 100)
# plt.plot(grid, (1 / np.pi) / (1 + grid ** 2))

plt.title('Banana distribution')
plt.ylabel(r'$y$')
plt.xlabel(r'$x$')

plt.show()
```

## Banana distribution

```python
logp_banana = lambda x: -x[:, 0] ** 2 / 200 - (x[:, 1] + 0.1 * x[:, 0] ** 2 - 100 * 0.1) ** 2 / 2
```

```python
test = UBVI(logp_banana, Gaussian(2), num_opt_steps=5000, n_samples=1000, n_init=500, init_inflation=100)
```

```python
test.build(10)
```

```python
samples = test._sample_g(100000).detach().numpy()

plt.figure(figsize=(10, 10))

plt.scatter(samples[:, 0], samples[:, 1], c=(0, 0, 1, 0.05))


x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
X, Y = np.meshgrid(x, y)
plt.contour(X, Y, -X ** 2 / 200 - (Y + 0.1 * X ** 2 - 100 * 0.1) ** 2 /2, [-10, -5, -1], colors='k')

plt.xlim([-30, 30])
plt.ylim([-30, 30])

# grid = np.linspace(-10, 10, 100)
# plt.plot(grid, (1 / np.pi) / (1 + grid ** 2))

plt.title('Boosting 2d gaussian distribution with nondiagonal covariance')
plt.ylabel(r'$y$')
plt.xlabel(r'$x$')

plt.show()
```

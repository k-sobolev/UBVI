```python
import torch

from matplotlib import pyplot as plt
import numpy as np

from gaussians import Gaussian
from ubvi import UBVI
```

```python
def logp_banana(X):
    b = 0.1
    x = X[:,0]
    y = X[:,1]
    return -x**2/200 - (y+b*x**2-100*b)**2/2 - np.log(2*np.pi*10)

def logp_cauchy(x):
    return -torch.log(1 + (x ** 2).sum(dim=-1))
example_log_pdf_1 = lambda x: -torch.log(1 + (x ** 2).sum(dim=-1))
```

```python
test = UBVI(logp_cauchy, Gaussian(1), num_opt_steps=1000, n_samples=100000, n_init=100, init_inflation=100)
```

```python
test.build(10)
```

```python
torch.log(torch.Teor(2*np.pi*10))
```

```python
res = test._hellsq_estimate()
```

```python
res.data.numpy().reshape((1))[0]
```

```python
hellinger_value = []
for i in range(1, 15):
    test.build(i)
    res = test._hellsq_estimate()
    hellinger_value.append(res.data.numpy().reshape((1))[0])
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = [i for i in range(1, 15)]
plt.plot(x, hellinger_value)
plt.xlabel('Number of components')
plt.ylabel('Hellinger distance')
plt.title("Hellinger distance vs number of components")
```
